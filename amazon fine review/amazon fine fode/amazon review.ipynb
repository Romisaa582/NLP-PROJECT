```python
df = df[['Text', 'Score']]
def classify_sentiment(score):
    if score <= 2:
        return 'Negative'
    elif score == 3:
        return 'Neutral'
    else:
        return 'Positive'

df['Sentiment'] = df['Score'].apply(classify_sentiment)
```

convert the text into lowercase


```python
df['Cleaned_Text'] = df['Text'].str.lower()
```

```python
df.head()
```

remove  punctuation

```python
import string

def remove_punctuation(text):
    return text.translate(str.maketrans("", "", string.punctuation))

df['Cleaned_Text'] = df['Cleaned_Text'].apply(remove_punctuation)
```

```python
df.head()
```

remove of stopwords

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    return " ".join(word for word in text.split() if word not in stop_words)

df['Cleaned_Text'] = df['Cleaned_Text'].apply(remove_stopwords)
```

remove_frequent_word

```python
from collections import Counter
word_counts = Counter(" ".join(df['Cleaned_Text']).split())
frequent_words = set([word for word, count in word_counts.items() if count > 100])

def remove_frequent_words(text):
    return " ".join(word for word in text.split() if word not in frequent_words)

df['Cleaned_Text'] = df['Cleaned_Text'].apply(remove_frequent_words)
df.head()
```

remove of rarewords

```python
rare_words = set([word for word, count in word_counts.items() if count == 1])

def remove_rare_words(text):
    return " ".join(word for word in text.split() if word not in rare_words)

df['Cleaned_Text'] = df['Cleaned_Text'].apply(remove_rare_words)
df
```

remove of special_characters

```python
import re

def remove_special_characters(text):
    return re.sub(r"[^a-zA-Z0-9\s]", "", text)

df['Cleaned_Text'] = df['Cleaned_Text'].apply(remove_special_characters)
df
```
